```python
# Import necessary libraries
import mne

# Path to your .bdf file
bdf_file_path = "sub-01/eeg/sub-01_task-targets_eeg.bdf"  # Replace with the actual path to your .bdf file

# Load the BDF file
raw = mne.io.read_raw_bdf(bdf_file_path, preload=True)

# Print channel names
channel_names = raw.info['ch_names']
print("Channel Names:", channel_names)

# Visualize the EEG data
raw.plot()
```

```python
# Plot the Power Spectral Density (PSD)
raw.plot_psd(fmax=50)  # Limit to 50 Hz for better visualization
```

```python
# Plot raw data with annotations
raw.plot(duration=10, n_channels=30, scalings='auto')  # Adjust duration and channels as needed
```
